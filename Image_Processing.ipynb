#**Libraries**

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```


#**Displaying Image From Drive**

#**Loading Pre-Trained Deep Learning Model**

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

#**Preprocessing The Test Image**

```python
test_image_path = "/content/drive/MyDrive/Final Assessment AI/Orange.jpg"

img = tf.keras.preprocessing.image.load_img(test_image_path, target_size = (224,224))

#Convert the Image into Array
img_arr = tf.keras.preprocessing.image.img_to_array(img)
print(img_arr.shape)

#For deep learning we need 1 more dimension
final_img = np.expand_dims(img_arr, axis=0)
final_img = tf.keras.applications.mobilenet.preprocess_input(final_img)

print(final_img.shape)
```

# **Testing the Image**

```python
predictions = mobile.predict(final_img)

results = tf.keras.applications.imagenet_utils.decode_predictions(predictions)
print(results)
```

# **Using Custom Model**

```python
#Created this model from https://teachablemachine.withgoogle.com

mobile = tf.keras.models.load_model('/content/drive/MyDrive/Final Assessment AI/keras_model.h5')
class_names = ['Orange', 'Banana', 'Papaya']

test_image_path = "/content/drive/MyDrive/Final Assessment AI/Banana.jpg"
img = tf.keras.preprocessing.image.load_img(test_image_path, target_size = (224,224))
#Convert the Image into Array
img_arr = tf.keras.preprocessing.image.img_to_array(img)
#For deep learning we need 1 more dimension
final_img = np.expand_dims(img_arr, axis=0)
final_img = tf.keras.applications.mobilenet.preprocess_input(final_img)
predictions = mobile.predict(final_img)
score = tf.nn.softmax(predictions[0])
print(
    "This image most likely belongs to {} with a {:.2f} percent confidence."
    .format(class_names[np.argmax(score)], 100 * np.max(score))
)
```
